```python
rows = 7

for i in range(rows):
    
    # L
    if i == rows-1:
        print("* "*5, end="  ")
    else:
        print("*", end="   ")

    # A
    for j in range(5):
        if (j==0 or j==4) and i!=0 or (i==0 and j>0 and j<4) or (i==3):
            print("*", end="")
        else:
            print(" ", end="")
    print("  ", end="")

    # V
    for j in range(5):
        if (j==0 and i<4) or (j==4 and i<4) or (i==4 and (j==1 or j==3)) or (i==5 and j==2):
            print("*", end="")
        else:
            print(" ", end="")
    print("  ", end="     ")

    # A
    for j in range(5):
        if (j==0 or j==4) and i!=0 or (i==0 and j>0 and j<4) or (i==3):
            print("*", end="")
        else:
            print(" ", end="")
    print("  ", end="")

    # N
    for j in range(5):
        if j==0 or j==4 or i==j:
            print("*", end="")
        else:
            print(" ", end="")
    print("  ", end="")

    # Y
    for j in range(5):
        if (i<3 and (i==j or i+j==4)) or (i>=3 and j==2):
            print("*", end="")
        else:
            print(" ", end="")
    print("  ", end="")

    # A
    for j in range(5):
        if (j==0 or j==4) and i!=0 or (i==0 and j>0 and j<4) or (i==3):
            print("*", end="")
        else:
            print(" ", end="")
    
    print()
```
